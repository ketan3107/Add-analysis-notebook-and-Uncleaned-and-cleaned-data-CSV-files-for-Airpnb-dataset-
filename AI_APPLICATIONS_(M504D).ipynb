```python
avg_price = df_cleaned.groupby('neighbourhood_group')['price'].mean()
print("Average Price by Neighborhood Group:\n", avg_price)
```

### Answer 1: It is clearly seen that 
### 'Manhattan 179.1 Dollors' has the highest average_price  
### 'Bronx - $85.3' has the best/lowest Price in all 5 neighbourhood_groups

# Question 3.2 :What type of room (private, shared, entire home) is most popular?

```python
room_popularity = df_cleaned.groupby(['neighbourhood_group', 'room_type']).size().unstack()
print("Room Type Popularity by Neighborhood Group:\n", room_popularity)
```

```python
room_popularity.plot(kind='bar', stacked=True, title="Room Type Popularity by Neighborhood Group")
plt.ylabel('Number of Listings')
plt.show()
```

# 3.3 How does host activity or the number of host listings relate to review frequency and guest satisfaction?

```python
host_review_freq = df_cleaned.groupby('calculated_host_listings_count')['reviews_per_month'].mean()
print("\n Average Reviews per Month by Host Listings Count:\n", host_review_freq)
```

```python
host_review_freq.plot(kind='line', marker='o', title='Host Listings Count vs Avg Reviews per Month')
plt.xlabel('Number of Listings per Host')
plt.ylabel('Average Reviews per Month')
plt.show()
```

## Answer 3.3 : It can be seen that Hosts with fewer than 10 listings generally get higher average reviews per month, indicating better guest engagement. Hosts with many listings tend to have lower average reviews per listing, suggesting it’s harder to maintain high engagement across a large portfolio.

# Question 3.5 :How do minimum stay requirements affect booking frequency and review patterns?

# Question 3.6 : What factors are most predictive of a listing's success (as measured by reviews per month or occupancy rates)?

```python
# Correlation matrix for factors predictive of listing success
corr = df_cleaned[['price', 'minimum_nights', 'number_of_reviews', 
                   'reviews_per_month', 'availability_365', 'calculated_host_listings_count']].corr()
print("\n Correlation Matrix:\n", corr)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Listing Features')
plt.show()
```

## Answer 3.6 - There is strong positive correlation between number_of_reviews and reviews_per_month (0.58), indicating listings with more reviews tend to get more frequent reviews monthly, a key success indicator.

## Availability_365 and calculated_host_listings_count show weak positive correlations with reviews_per_month (~0.17 and ~-0.05 respectively), suggesting they slightly influence listing success

## Minimum_nights has a moderate negative correlation (-0.21) with reviews_per_month, indicating longer minimum stays often reduce review frequency.

## Price shows almost no correlation with reviews_per_month (-0.05), implying price alone is not a strong predictor of review frequency.

## The most important factor linked to listing success (monthly reviews) is having a higher number of reviews overall. Listings with shorter minimum stay requirements also tend to perform better. Price and host listing count have little impact on review frequency.This means building positive review history and flexible stay options are key to Airbnb listing success in this dataset.
